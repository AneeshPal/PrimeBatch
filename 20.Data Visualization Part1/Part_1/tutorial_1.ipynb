```python
# Real Data
# Oscar Winner Data

oscar_movies=[
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years=[2008,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,232] # in $M

#Plot
plt.plot(years,oscar_revenue)

plt.title("Oscar Movies in each year")
plt.xlabel("Years")
plt.ylabel("Revenues (in $M)")
```

```python
# Non-Oscar Winner Movies
non_oscar_movies=["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue=[378,2788,829,185,275] #in $M

plt.plot(years,oscar_revenue,label="Oscar Movies")
plt.plot(years,non_oscar_revenue,label="Non-Oscar Movies")

plt.title("Oscar Movies V/s Non-Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenues (in $M)")
plt.legend()


```

```python
# Format String:consists of a part for color,marker and line.

# linewidth=5
#  marker="."
# color="blue"
# linestyle="-."

plt.plot(years,oscar_revenue,"<-y",label="Oscar Movies")
plt.plot(years,non_oscar_revenue,color="blue", marker=".",linestyle="-.",linewidth=5,label="Non-Oscar Movies")
```

```python
# Bar Chart 

oscar_movies=[
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years=[2008,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,232] # in $M

plt.bar(years,oscar_revenue,color="green")

plt.title("Revenue in each year for oscar movies")
plt.xlabel("Years")
plt.ylabel("Revenues (in $M)")


for i in range(len(years)):
    plt.text(years[i],oscar_revenue[i]+20,str(oscar_revenue[i]),ha="center")


plt.ylim(0,max(oscar_revenue)+100)
```

```python
# Compare two dataset using bar chart

oscar_movies=[
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years=[2008,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,232] # in $M

non_oscar_movies=["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue=[378,2788,829,185,275] #in $M

# plt.bar(years,oscar_revenue)
# plt.bar(years,non_oscar_revenue)

# plt.title("Oscar V/s Non-Oscar Movie Revenue")
# plt.xlabel("Year")
# plt.ylabel("Revenue ($m)")

# here both plot bar will overlap due to same graph width so we have to manipulate them to overcome this problem:
import numpy as np 
x=np.arange(len(years))
width=0.4

plt.bar(x-width/2,oscar_revenue,width,label="Oscar Movies")
plt.bar(x+width/2,non_oscar_revenue,width,label="Non Oscar Movies")

plt.title("Oscar V/s Non-Oscar Movie Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue ($m)")
plt.legend()

# but here our x-axis values get distorted they became index :
plt.xticks(x,years)
```

```python
# Scatter Plot:
# Used for:
# 1. outlier detection
# 2. clustering
# 3. Corelation

people=["Person A","Person B","Person C","Person D","Person E",
        "Person F","Person G","Person H","Person I","Person J"]
age=[22,25,30,35,40,45,50,55,60,65]
bp=[110,115,120,122,125,130,135,123,145,150]

# plot to find how bp changes with age:

# plt.scatter(age,bp)
# plt.title("Age V/s Bp")
# plt.xlabel("Age")
# plt.ylabel("Blood Pressure")
# plt.grid()

# Customisation on Scatter Plot:
# 1.Color
# 2.Size
# 3.Marker
# 4.Transparency: 0 to 1
# 5. Colormap

# colors=["green" if x<135 else "red" for x in bp]
# plt.scatter(age,bp,color=colors,s=bp,marker=">",alpha=0.2)
# plt.title("Age V/s Bp")
# plt.xlabel("Age")
# plt.ylabel("Blood Pressure")
# plt.grid()

plt.scatter(age,bp,s=bp,cmap="plasma",c=bp) #demonstration of colormap usuage
plt.title("Age V/s Bp")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()

# to know what color is representing what exactly i.e. what is the scale of color

# plt.colorbar()

plt.colorbar(label="Bp") # to add label to colorbar

# Annotations: which value is representing what on scatter plot

for i in range(len(people)):
    plt.annotate(people[i],xy=(age[i],bp[i]),xytext=(age[i]+1,bp[i]+1))

plt.xlim(min(age),max(age)+10)
plt.ylim(min(bp),max(bp)+5)
```
