```python
# Real Data
# Oscar Winner Data

oscar_movies=[
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years=[2008,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,232] # in $M

#Plot
plt.plot(years,oscar_revenue)

plt.title("Oscar Movies in each year")
plt.xlabel("Years")
plt.ylabel("Revenues (in $M)")
```

```python
# Non-Oscar Winner Movies
non_oscar_movies=["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue=[378,2788,829,185,275] #in $M

plt.plot(years,oscar_revenue,label="Oscar Movies")
plt.plot(years,non_oscar_revenue,label="Non-Oscar Movies")

plt.title("Oscar Movies V/s Non-Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenues (in $M)")
plt.legend()


```

```python
# Format String:consists of a part for color,marker and line.

# linewidth=5
#  marker="."
# color="blue"
# linestyle="-."

plt.plot(years,oscar_revenue,"<-y",label="Oscar Movies")
plt.plot(years,non_oscar_revenue,color="blue", marker=".",linestyle="-.",linewidth=5,label="Non-Oscar Movies")
```

```python
# Some Other Attributes
# 1. style.available
# 2. style.use()
# 3. grid()
# 4. tight_layout()
# 5. savefig("abc.png")


non_oscar_movies=["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue=[378,2788,829,185,275] #in $M
plt.plot(years,oscar_revenue,label="Oscar Movies")
plt.plot(years,non_oscar_revenue,label="Non-Oscar Movies")

plt.title("Oscar Movies V/s Non-Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenues (in $M)")

plt.style.available
plt.style.use("ggplot")
plt.grid()
plt.legend()
plt.show()
    
```

```python
# tight_layout() usuage:

oscar_movies=[
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years=[2008,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,232] # in $M

#Plot
plt.plot(oscar_movies,oscar_revenue)

plt.title("Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenues (in $M)")

plt.tight_layout()

plt.savefig("final_plot.png")
```
