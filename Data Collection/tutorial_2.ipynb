```python
import requests

URL="https://www.scrapethissite.com/pages/simple/"

res=requests.get(URL)
```

```python
res
```

```python
if res.status_code==200:
    print(res.headers)

with open("scraped_data/data1.html","w") as f:
    f.write(res.text)
```

```python
# BeautifulSoup Library & lxml

!pip install beautifulsoup4
!pip install lxml

from bs4 import BeautifulSoup

with open("scraped_data/data1.html","r") as f:
    html_content=f.read()

soup=BeautifulSoup(html_content,"lxml")
```

```python
 soup
```

```python
# trying to extract country name and its corresponding population

all_h3=soup.find_all("h3")

all_countries=[]

for h3 in all_h3:
    name=h3.get_text(strip=True)

    population=h3.find_next("div").select("span.country-population")[0].get_text(strip=True)

    all_countries.append([name,population])
```

```python
# all_countries

import pandas as pd

df=pd.DataFrame(all_countries,columns=["Name","Population"])
```

```python
df
```

```python
df.to_csv("cleaned_data/data.csv",index=False)
```
